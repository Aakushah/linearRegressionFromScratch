```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Dummy Data
years_experience = np.array([1.1, 2.3, 3.0, 3.8, 4.5, 5.2, 6.0, 6.8, 7.5, 8.2, 9.0, 10.0])
salary = np.array([22000, 25000, 29000, 37000, 41000, 48000, 55000, 62000, 69000, 75000, 82000, 90000])

# Create DataFrame
df = pd.DataFrame({
    'YearsExperience': years_experience,
    'Salary': salary
})

# Step 2: Save Data to CSV
df.to_csv('data_LinearRegression.csv', index=False)

# Step 3: Load Data from CSV
df = pd.read_csv('data_LinearRegression.csv')

# Step 4: Reshape Data for Matrix Operations
x = np.array(df.iloc[:, 0]).reshape(-1, 1)  # YearsExperience as feature
y = np.array(df.iloc[:, 1]).reshape(-1, 1)  # Salary as target


# Step 5: Initialize Parameters for Linear Regression (m and b)
m = 0  # Slope (initial)
b = 0  # Intercept (initial)
learning_rate = 0.01  # Learning rate for gradient descent
epochs = 1000  # Number of iterations

# Step 6: Implementing Gradient Descent
n = len(x)  # Number of samples

for epoch in range(epochs):
    # Predicted Salary using current m and b
    y_pred = m * x + b


    # Compute the error
    error = y_pred - y


    # Compute gradients for m (slope) and b (intercept)
    dm = (2/n) * np.sum(x * error)
    db = (2/n) * np.sum(error)

    # Update parameters using the gradients
    m -= learning_rate * dm
    b -= learning_rate * db

    # Print loss every 100 epochs for tracking progress
    # if epoch % 100 == 0:
    #     loss = np.mean(error ** 2)  # Mean Squared Error
    #     print(f"Epoch {epoch}: Loss = {loss:.4f}")

# Step 7: Final Slope and Intercept after Gradient Descent
print(f"\nFinal Slope (m): {m:.4f}")
print(f"Final Intercept (b): {b:.4f}")

# Step 8: Predict Salary using the trained model
y_pred = m * x + b
print('see the differences of Predicted value and Actual value')
print(np.column_stack((y_pred,y)))

# Step 9: Plot the Results (Actual vs Predicted)
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.title('Linear Regression from Scratch')
plt.show()





```
